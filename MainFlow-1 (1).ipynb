```python
# Import necessary libraries for outlier detection and handling
import numpy as np
from scipy.stats import zscore

# Calculate the z-scores for numerical columns
z_scores = np.abs(zscore(df.select_dtypes(include='number')))

# Set a threshold for outliers (e.g., z-score greater than 3)
outlier_threshold = 3

# Identify rows with outliers
outliers = np.where(z_scores > outlier_threshold)

# Remove rows with outliers
df_without_outliers = df[(z_scores < outlier_threshold).all(axis=1)]

# Display the number of rows removed due to outliers
num_outliers_removed = len(df) - len(df_without_outliers)
print("Number of rows with outliers removed:", num_outliers_removed)
```

```python
# Identify rows with outliers
outliers = np.where(z_scores > outlier_threshold)

# Display the rows that were outliers and are removed
outliers_removed_data = df.iloc[outliers[0]]
print("Outliers that were removed:")
print(outliers_removed_data)
```

```python
# Extract the columns with 'Observation_' prefix for one-hot encoding
observation_columns = [col for col in df.columns if col.startswith('Observation_')]

# One-hot encode the 'Observation' columns
df = pd.get_dummies(df, columns=observation_columns)

# Display the transformed dataset
print("Transformed Dataset:")
print(df.head())
```

```python
# Remove duplicates
df.drop_duplicates(inplace=True)

# Detect and address outliers (replace with an appropriate method)
# For demonstration, we're assuming outliers have been replaced with mean
# Replace this with your outlier detection and handling method
df_outliers_removed = df.clip(lower=df.mean() - 3*df.std(), upper=df.mean() + 3*df.std(), axis=1)

# Create a new dataset without missing values, duplicates, and outliers
cleaned_dataset = df_outliers_removed.dropna()

# Display the cleaned dataset
print("Cleaned Dataset:")
print(cleaned_dataset.head())
```
