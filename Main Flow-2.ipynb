```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
data = pd.read_csv(r'C:\Users\harsh\Downloads\Main Flow Internship\INDIAvi.csv')
```

```python
# Display first few rows and summary statistics
print(data.head())
print(data.describe())
```

```python
# Data Cleaning
# Handle missing values
data.fillna(method='ffill', inplace=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histogram for views
plt.figure(figsize=(10, 6))
sns.histplot(data=data, x='views', bins=30, kde=True)
plt.title('Distribution of Views')
plt.xlabel('Views')
plt.ylabel('Frequency')
plt.show()

# Scatter plot for likes vs. dislikes
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='likes', y='dislikes')
plt.title('Likes vs. Dislikes')
plt.xlabel('Likes')
plt.ylabel('Dislikes')
plt.show()

# Box plot for comment_count
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, y='comment_count')
plt.title('Box Plot of Comment Count')
plt.ylabel('Comment Count')
plt.show()
```

```python
# Data Visualization
# Visualize data distributions using histograms
for col in data.columns:
    if data[col].dtype == 'float64' or data[col].dtype == 'int64':
        plt.figure(figsize=(8, 6))
        sns.histplot(data[col], kde=True)
        plt.title(f'Distribution of {col}')
        plt.xlabel(col)
        plt.ylabel('Count')
        plt.show()
```

```python
# Select numerical columns for pair plot
numerical_columns = ['views', 'likes', 'dislikes', 'comment_count']

# Create pair plot for numerical columns
sns.pairplot(data=data[numerical_columns])
plt.show()
```

```python
correlation_matrix = data.corr(numeric_only=True)
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

```python
column_names = data.columns.tolist()
print(column_names)
```

```python
# Define numerical columns for summary statistics
numerical_columns = ['views', 'likes', 'dislikes', 'comment_count']

# Group by 'category_id' and calculate summary statistics
grouped_data = data.groupby('category_id')[numerical_columns].agg(['mean', 'median', 'count'])
print(grouped_data)
```

```python
# Box plot for 'category_id' vs numerical variables
for col in numerical_columns:
    plt.figure(figsize=(12, 8))
    sns.boxplot(x='category_id', y=col, data=data)
    plt.title(f'Box Plot of {col} by category_id')
    plt.show()
```

Based on the exploratory data analysis (EDA) and the initial insights obtained from the dataset, we can formulate several hypotheses for further analysis. These hypotheses can guide more in-depth investigations and testing to gain a deeper understanding of the data and potential relationships. Here are some hypotheses:

1. Hypothesis 1: Engagement and Video Popularity:
   - Null Hypothesis (H0): The number of likes, dislikes, and comments on a video is not significantly associated with its popularity (measured by views).
   - Alternative Hypothesis (H1): The number of likes, dislikes, and comments on a video is significantly associated with its popularity (measured by views).

2. Hypothesis 2: Category Influence on Popularity:
   - Null Hypothesis (H0): The category of the video does not influence its popularity (measured by views).
   - Alternative Hypothesis (H1): The category of the video significantly influences its popularity (measured by views).

3. Hypothesis 3: Comments and Likes Correlation:
   - Null Hypothesis (H0): There is no significant correlation between the number of comments and the number of likes on a video.
   - Alternative Hypothesis (H1): There is a significant correlation between the number of comments and the number of likes on a video.

4. Hypothesis 4: Engagement and Dislikes:
   - Null Hypothesis (H0): The number of dislikes is not significantly associated with the engagement level of a video (likes, comments).
   - Alternative Hypothesis (H1): The number of dislikes is significantly associated with the engagement level of a video (likes, comments).

5. Hypothesis 5: Publishing Time and Popularity:
   - Null Hypothesis (H0): The time of publishing (e.g., day of the week, time of day) does not influence the popularity of a video.
   - Alternative Hypothesis (H1): The time of publishing significantly influences the popularity of a video.
